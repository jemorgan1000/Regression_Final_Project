# Predicting Death Precentage for Coronavirus

```python
import os.path as osp
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = osp.dirname(osp.abspath(os.getcwd()))
data_path = osp.abspath(osp.join(path,'Data'))
county_path_2 = osp.join(data_path,'County_Dems_Processed_2.csv')
```

```python
jhu_df = pd.read_csv(osp.join(data_path,'county_data_jhu_4_8.csv'))
jhu_df = jhu_df[jhu_df.Country_Region == "US"]
remove_provs = ['Puerto Rico','Grand Princess','Guam','Diamond Princess','Virgin Islands','Northern Mariana Islands']
jhu_df = jhu_df[~(jhu_df.Province_State.isin(remove_provs))]
jhu_df.insert(len(jhu_df.columns),'Percent',jhu_df.Deaths/jhu_df.Confirmed)
jhu_df.head()
```

```python
jhu_df.Confirmed.plot.hist()
```

```python
%matplotlib inline
fig, ax = plt.subplots()
ax.scatter(jhu_df.Confirmed,jhu_df.Deaths)
```

```python
jhu_df = jhu_df[jhu_df.Confirmed >= 200]
jhu_df.Percent.plot.hist()
```

```python
%matplotlib inline
fig, ax = plt.subplots()
ax.scatter(jhu_df.Confirmed,jhu_df.Percent)
```

```python
jhu_df_2 = jhu_df[jhu_df.Confirmed >= 100]
%matplotlib inline
fig, ax = plt.subplots()
ax.scatter(jhu_df_2.Confirmed,jhu_df_2.Percent)
```

```python
county_df = pd.read_csv(county_path_2,index_col=0)
county_df.head()
```

```python
county_df.columns
```

```python
# make FIPS column for join
county_df.insert(len(county_df.columns),"FIPS", county_df.state * 1000 + county_df.county)

# make pop columns, under 20, 20-35, 35-45, 45-55, 55-65, 65-75, 75+
county_df.insert(len(county_df.columns),'POP_UNDER_20',county_df.POP_UNDER_5 + county_df.POP_5_9 + county_df.POP_10_14 + county_df.POP_15_19)
county_df.insert(len(county_df.columns),'POP_20_35',county_df.POP_20_24 + county_df.POP_25_34)
county_df.insert(len(county_df.columns),'POP_55_65',county_df.POP_55_59 + county_df.POP_60_65)
county_df.insert(len(county_df.columns),'POP_75_UP',county_df.POP_75_84 + county_df.POP_85_P)

# make income columns UNder 35, 35_50, 50_75, 75_100, INC_100_150, INC_150_P
county_df.insert(len(county_df.columns),'INC_UNDER_35',county_df.INC_U_10 + county_df.INC_10_15 + county_df.INC_15_25 + county_df.INC_25_34)
county_df.insert(len(county_df.columns),'INC_150_UP',county_df.INC_150_200 + county_df.INC_200_G)
county_df.columns

```

```python
comp_df = county_df.merge(jhu_df_2,on="FIPS")
print(comp_df.columns)
comp_df.head()
```

```python
ts_df = pd.read_csv(osp.join(data_path,'time_series_covid19_confirmed_US.csv'))
ts_df = ts_df.drop(['UID','iso2','iso3','code3','Admin2',"Province_State",'Country_Region','Lat','Long_','Combined_Key'],axis=1)
ts_df.dropna(subset=['FIPS'],inplace=True)
ts_df = ts_df.astype({'FIPS':'int32'})
ts_df.set_index(['FIPS'],inplace=True)
ts_df = ts_df.T
ts_df.index = pd.to_datetime(ts_df.index)
days_df = pd.DataFrame([ts_df.applymap(lambda x: x >= 500).sum(), ts_df.applymap(lambda x: x >= 100).sum()]).T
days_df.columns = ['Days_Since_500','Days_Since_100']
days_df = days_df.reset_index()
print(ts_df.head())
print(days_df.head())
```

```python
comp_df = comp_df.merge(days_df,on='FIPS')
```

```python
data_df = comp_df[["WHITE_POP","TOTAL_POP",'BLACK_POP','HOUSING_UNITS',
                   'POP_UNDER_20','POP_25_34','POP_35_44','POP_45_54','POP_55_65','POP_65_74','POP_75_UP','INC_UNDER_35','INC_34_49',
                   'INC_75_100','INC_100_150','INC_150_UP','Days_Since_500','Days_Since_100','Confirmed','Percent']]
data_df = data_df.astype('float32')
data_df.WHITE_POP = data_df.WHITE_POP /100
data_df.BLACK_POP = data_df.BLACK_POP /100
data_df.Percent = data_df.Percent * 100
data_df.head()
```

```python
data_df.describe().to_csv('America_Data_Desc.csv')
```

```python
from sklearn import preprocessing
from sklearn import linear_model
from sklearn import decomposition
from sklearn.metrics import r2_score
```

```python
pca = decomposition.PCA(n_components=2)
pca.fit(data_df)
X = pca.transform(data_df)
fig, ax = plt.subplots()
ax.scatter(X[:,0],X[:,1])
```

```python
y = data_df.Percent
X = data_df[["WHITE_POP","TOTAL_POP",'BLACK_POP','HOUSING_UNITS',
                   'POP_UNDER_20','POP_25_34','POP_35_44','POP_45_54','POP_55_65','POP_75_UP','INC_UNDER_35','INC_34_49',
                   'INC_75_100','INC_100_150','INC_150_UP','Confirmed']]
reg = linear_model.Ridge(alpha=0.1)
reg.fit(X,y)
reg.predict(X)
r2_score(reg.predict(X),y)
```
