### Georgia Risk Factor Model 
To build a complete risk factor model with chronic conditions we will need to make a series of assumptions. 
#### Assumption 1
We assume that people get the disease at Random and thus any sample of individuals with the disease show the characteristics of the overal population. We can relax this assumption in a data generation process where we assume that the population that has been tested or addmitted is sicker. We should be able to play with this number to test for robustness of the data generation procedure.
#### Assumption 2
At this point in time,the healthcare system has not been taxed thus we assume that hospital care is the same for all patients.
#### Assumption 3
We can use state of Georgia Data to determine the conditional probability of having a specific chronic condition of interest. We could use data from the American Community Survey to further refine these estimates. We will then assign a probablity to each individual in each county of having a specific condition. This will give us variability in each county of the state.

#### Notes
Unfortunately, we do not know the number hospitalized from each county and where they are being hospitalized. This prohibits us from using hospitilization as an adverse outcome or the use of an ICU bed. Acute Respiratory Distress syndrome (ARDS) is a negative outcome that creates scarring in the lungs making it harder to breathe. This outcome is averse requires months of physical therapy.

### Data Preprocessing
We will take data from the American Community Survey. Do to a large majority of cases in the metro atlanta area. We will use Fulton, dekalb, and clayton counties as our data points. 

```python
census_path = osp.join(data_path, 'census_2010.csv')
census_df = pd.read_csv(census_path)
census_df = census_df.groupby("County").sum()
census_df.reset_index(inplace=True)
drop_cols = [
       'Area (Land)', 'Area (Water)', 'Summary Level', 'Geographic Component',
       'Region', 'Division', 'FIPS', 'State (FIPS)', 'Census Tract',"Block", 'Total Population.1',]
census_df.drop(columns = drop_cols,inplace=True)
census_df.head()
```

```python
drop_cols_2 = ['Total Population: Male: Under 5 years',
       'Total Population: Male: 5 to 9 years',
       'Total Population: Male: 10 to 14 years',
       'Total Population: Male: 15 to 17 years','Total Population: Female: Under 5 years',
       'Total Population: Female: 5 to 9 years',
       'Total Population: Female: 10 to 14 years',
       'Total Population: Female: 15 to 17 years','Households',
       'Households: Family households',
       'Households: Family households: Married-couple family',
       'Households: Family households: Other family',
       'Households: Family households: Other family: Male householder, no wife present',
       'Households: Family households: Other family: Female householder, no husband present',
       'Households: Nonfamily households',
       'Households: Nonfamily households: Householder living alone',
       'Households: Nonfamily households: Householder not living alone','Total Population: 5 to 9 years',
       'Total Population: 10 to 14 years', 'Total Population: 15 to 17 years']
census_df.drop(columns = drop_cols_2,inplace=True)
census_df.head()
```

```python
county_map_df = pd.read_csv(osp.join(data_path,'fips_2_county.csv'))
county_map_df = county_map_df[county_map_df.state == "GA"]
county_map_df.name = county_map_df.name.str.replace("County",'')
county_map_df.name = county_map_df.name.str.strip()
county_map_df.fips = county_map_df.fips -13000
county_map_df.head()
```
