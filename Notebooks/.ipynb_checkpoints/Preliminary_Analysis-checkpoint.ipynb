```python
ga_death_path = osp.join(data_path, 'georgia_covid_deaths_4_7_19.csv')
ga_death_path
```

```python
ga_death_df = pd.read_csv(ga_death_path)
ga_death_df.head()
```

```python
def check_chronic(x):
    if x == "Yes":
        return 1
    elif x == "No":
        return 0
    else:
        return -1
```

```python
ga_death_df.insert(len(ga_death_df.columns),'Gender_IND',ga_death_df.Gender.apply(lambda x: (x== "MALE")*1))
ga_death_df.insert(len(ga_death_df.columns),'Underlying_IND',ga_death_df.Underlying.apply(check_chronic))
ga_death_df.head()
```

### Georgia Risk Factor Model 
To build a complete risk factor model with chronic conditions we will need to make a series of assumptions. 
#### Assumption 1
We assume that people get the disease at Random and thus any sample of individuals with the disease show the characteristics of the overal population. We can relax this assumption in a data generation process where we assume that the population that has been tested or addmitted is sicker. We should be able to play with this number to test for robustness of the data generation procedure.
#### Assumption 2
At this point in time,the healthcare system has not been taxed thus we assume that hospital care is the same for all patients.
#### Assumption 3
We can use state of Georgia Data to determine the conditional probability of having a specific chronic condition of interest. We could use data from the American Community Survey to further refine these estimates. We will then assign a probablity to each individual in each county of having a specific condition. This will give us variability in each county of the state.

#### Notes
Unfortunately, we do not know the number hospitalized from each county and where they are being hospitalized. This prohibits us from using hospitilization as an adverse outcome or the use of an ICU bed.
